```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
cancer_drug_df = (pd.merge(mouse_metadata, study_results, how = "outer", on = ['Mouse ID']))

# Display the data table for preview
cancer_drug_df.head()
```

```python
# Checking the number of mice.
cancer_drug_df['Mouse ID'].nunique()
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mice = cancer_drug_df.loc[cancer_drug_df.duplicated(subset=['Mouse ID', 'Timepoint']), 'Mouse ID'].unique()
duplicate_mice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mice_data = cancer_drug_df.loc[cancer_drug_df['Mouse ID'] == 'g989', :]
duplicate_mice_data
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cancer_drug_df_cleaned = cancer_drug_df[cancer_drug_df['Mouse ID'].isin(duplicate_mice)==False]
cancer_drug_df_cleaned.head()
```

```python
# Checking the number of mice in the clean DataFrame.
cancer_drug_df_cleaned['Mouse ID'].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
tumor_regimen_mean = cancer_drug_df_cleaned['Tumor Volume (mm3)'].groupby(cancer_drug_df_cleaned['Drug Regimen']).mean()
tumor_regimen_median = cancer_drug_df_cleaned['Tumor Volume (mm3)'].groupby(cancer_drug_df_cleaned['Drug Regimen']).median()
tumor_regimen_variance = cancer_drug_df_cleaned['Tumor Volume (mm3)'].groupby(cancer_drug_df_cleaned['Drug Regimen']).var()
tumor_regimen_std = cancer_drug_df_cleaned['Tumor Volume (mm3)'].groupby(cancer_drug_df_cleaned['Drug Regimen']).mean().std()
tumor_regimen_sem = cancer_drug_df_cleaned['Tumor Volume (mm3)'].groupby(cancer_drug_df_cleaned['Drug Regimen']).sem()

tumor_regimen_stats_df = pd.DataFrame({"Mean Tumor Volume": tumor_regimen_mean,
                                      "Median Tumor Volume": tumor_regimen_median,
                                      "Tumor Volume Variance": tumor_regimen_variance,
                                      "Tumor Volume Std. Dev.": tumor_regimen_std,
                                      "Tumor Volume Std. Err.": tumor_regimen_sem})
tumor_regimen_stats_df
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
tumor_regimen_stats_summ = cancer_drug_df_cleaned[["Drug Regimen","Tumor Volume (mm3)"]].groupby(["Drug Regimen"]).agg(['mean', 'median', 'var', 'std', 'sem'])
tumor_regimen_stats_summ
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
# Put treatments into a list for for loop (and later for plot labels
drug_types = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
drug_group = cancer_drug_df_cleaned[cancer_drug_df_cleaned['Drug Regimen'].isin(drug_types)]

# Start by getting the last (greatest) timepoint for each mouse
# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
last_timepoint = drug_group.groupby(["Drug Regimen", "Mouse ID"]).agg(tum_vol=("Tumor Volume (mm3)", lambda x: x.iloc[-1]))
last_timepoint = last_timepoint.stack(level=0).unstack(level=0)

drug_group.head()
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tum_vol_data = 0

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Locate the rows which contain mice on each drug and get the tumor volumes
for drug in drug_types:
    quartiles = last_timepoint[drug].quantile([.25, .5, .75]).round(2)
    lowerq = quartiles[0.25].round(2)
    upperq = quartiles[0.75].round(2)
    iqr = round(upperq-lowerq,2)
    lowerbound = round(lowerq - (1.5*iqr),2)
    upperbound = round(upperq + (1.5*iqr),2)
    
    # Determine outliers using upper and lower bounds
    if tum_vol_data == 0:
        print(f"")
    print(f"The lower quartile of {drug} treatments is: {lowerq}")
    print(f"The upper quartile of {drug} treatments is: {upperq}")
    print(f"The IQR of {drug} treatments is: {iqr}")
    print(f"Values below {lowerbound} could be outliers")
    print(f"Values above {upperbound} could be outliers")
    print(f"")    
    
    tum_vol_data += 1
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
tum_vol_list = []
for drug in drug_types:
    tum_vol_list.append(list(last_timepoint[drug].dropna()))

fig1, ax1 = plt.subplots()
ax1.set_title("Tumor Volume for each Treatment type")
ax1.set_ylabel("Final Tumor Volume (mm3)")
ax1.boxplot(tum_vol_list, flierprops = {'marker': 'o', 'markersize': 12, 'markerfacecolor': 'red'})
plt.xticks([1,2,3,4],["Capomulin", "Ramicane", "Infubinol", "Ceftamin"])
plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a single mouse treated with Capomulin
capomulin_l509 = cancer_drug_df_cleaned.loc[cancer_drug_df_cleaned['Drug Regimen'] == 'Capomulin',:]
mouse_l509 = capomulin_l509.loc[capomulin_l509['Mouse ID'] == "l509",:]
x_axis = mouse_l509['Timepoint']
tum_vol_per_l509 = mouse_l509["Tumor Volume (mm3)"]

plt.title("Capomulin treatment of mouse l509")
plt.plot(x_axis, tum_vol_per_l509,)
plt.xlabel("Timepoint (days)")
plt.ylabel("Tumor Volume (mm3)")
plt.show()
```

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
mouse_tum_wei = capomulin_l509.groupby(['Mouse ID']).mean()

plt.scatter(mouse_tum_wei['Weight (g)'], mouse_tum_wei["Tumor Volume (mm3)"])
plt.title("Mouse weight versus Avg Tumor Volume Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen
correlation = st.pearsonr(mouse_tum_wei['Weight (g)'], mouse_tum_wei["Tumor Volume (mm3)"])
print(f"The correlation between mouse weight and the average tumor volume is {round(correlation[0],2)}")

wt_slope, wt_int, wt_r, wt_p, wt_std_err = st.linregress(mouse_tum_wei['Weight (g)'], mouse_tum_wei["Tumor Volume (mm3)"])
wt_fit = wt_slope * mouse_tum_wei['Weight (g)'] + wt_int

plt.scatter(mouse_tum_wei['Weight (g)'], mouse_tum_wei["Tumor Volume (mm3)"])
plt.title("Mouse weight versus Avg Tumor Volume Capomulin")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.plot(mouse_tum_wei['Weight (g)'], wt_fit, color='r')
plt.show

```
